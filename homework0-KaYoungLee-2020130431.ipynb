```python
# Problem 1. Make a calculator that can do the following operations
def Function (num1, num2, operator):
    if operator == '+':
        answer = num1 + num2
    elif operator == '-':
        answer = num1 - num2
    elif operator == '*':
        answer = num1 * num2
    elif operator == '/':
        answer = num1 / num2
    elif operator == '%':
        answer = num1 % num2
    return answer

print(Function(4, 2, '+')) #6
print(Function(4, 2, '-')) #2
print(Function(4, 2, '*')) #8
print(Function(4, 2, '/')) #2.0
print(Function(4, 2, '%')) #03
```

```python
# Problem 2. Countring Frequency
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# A. Make a 10x10 sized array filled with integers [0-9] by using numpy.
arr = np.random.randint(low=0, high=10, size=(10,10))
print(arr)

# B. Count the number of appearances of each integer, and organize it as 'dataframe' or 'series' format by using pandas.
unique, count = np.unique(arr, return_counts=True)
df = pd.DataFrame(count, unique)
print(df)

# C. Make a histogram of frequency of each number by using matplotlib
arr2 = np.concatenate(arr)
print(arr2)
plt.hist(arr2)
plt.show()
```

```python
# Problem 3. Sorting Problem
arr = np.random.randint(low = 0, high = 10000, size=1000)
print(arr)

sorting_array = [10001]
for i in arr:
    for j in range(len(sorting_array)):
        if i <= sorting_array[j]:
            sorting_array.insert(j, i)
            break
del sorting_array[1000]

print(sorting_array)
print(len(sorting_array))
```
